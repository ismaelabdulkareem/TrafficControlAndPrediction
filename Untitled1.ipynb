```python
def count_files_in_folder(folder_path):
    # List all files and directories in the specified folder
    files_and_dirs = os.listdir(folder_path)

    # Filter out only the files
    files = [f for f in files_and_dirs if os.path.isfile(os.path.join(folder_path, f))]

    # Return the count of files
    return len(files)

# Specify the path to your folder
folder_path = '/content/drive/MyDrive/datasetmeEx/datasetme'

# Count the files in the folder
num_files = count_files_in_folder(folder_path)

print(f'There are {num_files} files in the folder "{folder_path}".')
```

```python
dataset_path = '/content/drive/MyDrive/datasetmeEx/datasetme'


race = []
image_paths = []

image_filenames = os.listdir(dataset_path)
random.shuffle(image_filenames)

for image in tqdm(image_filenames):
  image_path = os.path.join(dataset_path, image)
  img_components = image.split('_')
  race_label = int(img_components[2])


  # Append the image_path, and race_label
  race.append(race_label)
  image_paths.append(image_path)
```

```python
# Create a DataFrame from the lists
data = {
    'image': image_paths,
    'race': race
}
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())
print(df.tail())
```

```python
print("Total number of images in the dataset: ", len(df))

label_count = df['race'].value_counts()
print(label_count)
```
